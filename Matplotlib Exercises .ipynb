## Exercise 1

** Follow along with these steps: **

* ** Use add_axes to add an axis to the figure canvas at [0,0,1,1]. Call this new axis ax. **

** Now plot (x,y) on both axes. And call your figure object to show it.**

```python
# create a blank canvas
fig = plt.figure()

ax1 = fig.add_axes([0,0,1,1]) # defineing blank figure
ax2 = fig.add_axes([0.2,0.5,.2,.2]) #taken form instructions

# creates a graph within
ax1.plot(x,y,'b') # 'b' will make the line blue
ax2.plot(x,y,'r') # 'r' will make the line red

#Titles
ax1.set_xlabel('X_label_axes1')
ax1.set_ylabel('Y_label_axes1')
ax1.set_title('Axes 1 Title')

ax2.set_xlabel('X_label_axes2')
ax2.set_ylabel('Y_label_axes2')
ax2.set_title('Axes 2 Title')
```

** Now use x,y, and z arrays to recreate the plot below. Notice the xlimits and y limits on the inserted plot:**

```python
fig = plt.figure()

ax1 = fig.add_axes([0,0,1,1]) # defineing blank figure
ax2 = fig.add_axes([0.2,0.5,.4,.4])

#adjust x and y limits
ax2.set_xlim(20.0,22.0)
ax2.set_ylim(30,50)
ax2.set_title('zoom')
ax2.set_xlabel('X')
ax2.set_ylabel('Y')
ax2.plot(x,y,'r')

ax1.set_xlim(0,100)
ax1.set_ylim(0,10000) # went ahead and manually adjust these, lookng back Z may have gone this high automatically
ax1.set_ylabel('Z') # notice in the example that we are plotting Z 
ax1.set_xlabel('X')
ax1.plot(x,z,'b') 
```
